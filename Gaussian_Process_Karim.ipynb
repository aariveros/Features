Generate samples from a GP with a Squared Exponential kernel

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def generate_SQE_LC(L,sigma, num_obs, range_min_t,range_max_t):

    Kernel = lambda x,y:sigma**2*np.exp( -0.5*( ((x-y)/L)**2 ) )# square exponential Kernel
    time = np.linspace(range_min_t, range_max_t, num_obs)
    mean_mag = 0.0#(range_max_t-range_min_t)/2.0
    Gram_Matrix = np.zeros((num_obs, num_obs))
    for i in range(0,num_obs):
        for j in range(0,num_obs):
            Gram_Matrix[i][j] = Kernel(time[i],time[j])
            #if i == j:
            #    Gram_Matrix[i][j] += np.random.normal(0.0, 1.0, 1)

    mag = np.random.multivariate_normal(np.ones(num_obs)*mean_mag, Gram_Matrix)
    error = np.random.uniform(0.0, 0.3, num_obs)

    return time, mag, error

num_LCs = 4
num_obs = 300
range_min_t = 0.0
range_max_t = 1.0
L_sqe = 0.1
sigma = 0.3
for i in range(0,num_LCs):
    time, mag, error = generate_SQE_LC(L_sqe, sigma, num_obs, range_min_t,range_max_t)
    plt.plot(time,mag)

#plt.show()


```

Generate samples from a GP with a polynomial kernel:

```python
import numpy as np
#import imp
#from LightCurves import *
#imp.load_source('LightCurves', './LightCurves.py')
import matplotlib.pyplot as plt
%matplotlib inline

def generate_polynomial_LC(L,P,sigma, num_obs, range_min_t,range_max_t):

    Kernel = lambda x,y:sigma**2*np.exp(-2.0*np.sin(np.pi*np.abs(x-y)/P)**2 / (L**2) )# Polynomial Kernel
    time = np.linspace(range_min_t, range_max_t, num_obs)
    mean_mag = 0.0#(range_max_t-range_min_t)/2.0
    Gram_Matrix = np.zeros((num_obs, num_obs))
    for i in range(0,num_obs):
        for j in range(0,num_obs):
            Gram_Matrix[i][j] = Kernel(time[i],time[j])
            #if i == j:
            #    Gram_Matrix[i][j] += np.random.normal(0.0, 0.1, 1)

    mag = np.random.multivariate_normal(np.ones(num_obs)*mean_mag, Gram_Matrix)
    error = np.random.uniform(0.0, 0.3, num_obs)

    return time, mag, error


num_LCs = 3
L = 0.7
num_obs = 300
range_min_t = 0.0
range_max_t = 1.0
#range_min_mag = 0.0
#range_max_mag = 1.0

P = (range_max_t-range_min_t)/5.0
sigma = 0.5

for i in range(0,num_LCs):
    time, mag, error = generate_polynomial_LC(L,P,sigma,num_obs, range_min_t,range_max_t)
    plt.plot(time,mag)

plt.show()

```
